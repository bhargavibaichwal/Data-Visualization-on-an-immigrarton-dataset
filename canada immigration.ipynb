## 14.Using the scripting layer and the df dataset, create a horizontal bar plot showing the total number of immigrants to Canada from the top 15 countries, for the period 1980 - 2013. Label each country with the total immigrant count.

## 19. Create a scatter plot of the total immigration from Denmark, Norway, and Sweden to Canada from 1980 to 2013?

## --visiualise for top 5 cotinents


```python
can.set_index('Continent',inplace=True)
```

```python
df_continents=can.groupby('Continent',axis=0).sum()# first group the continents
df_continents.head()
```

```python
# now calling the column
df_continents['Total'].plot(kind='pie',
                           figsize=(7,6),
                           autopct='%1.2f%%',# show the decimal point
                            startangle=90,
                           shadow=True)
```

```python
# for slicing the pie plot 
colors_list=['gold','yellowgreen','lightcoral','lightskyblue','lightgreen','red']
explode_list=[0.3,0,0,0,0,0]# ratio for each continent
df_continents['Total'].plot(kind='pie',
                           figsize=(6,6),
                           autopct='%1.2f%%',# show the decimal point
                            startangle=90,
                           shadow=True,
                           labels=None,# removed the labels
                           colors=colors_list,
                           explode=explode_list,
                           pctdistance=0.8)# changes the place of the  label
```
